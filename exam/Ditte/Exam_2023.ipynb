```r
# fit model
church_model <- lm(church ~ H3020 + H4058 + H5020 + H8090, data = exam2023)

# extract values
rse <- summary(church_model)$sigma        # residual standard error
df  <- church_model$df.residual           # residual degrees of freedom
SSR <- rse^2 * df                          # sum of squared residuals

y    <- exam2023$church
SST  <- sum((y - mean(y))^2)               # total sum of squares

R2_manual <- 1 - SSR / SST                 # manual R^2
R2_model  <- summary(church_model)$r.squared

cat("Manual R2:", R2_manual, "\n")
cat("Model R2 :", R2_model, "\n")
```

R² is

$$ R^2 = 1 - \frac{SSR}{SST} $$

where:

* $SSR$ is the sum of squared residuals
* $SST$ is the total sum of squares

From the output:

* Residual standard error $RSE = 3072$
* Degrees of freedom for residuals $df = 93$

**Step 1**

Compute the sum of squared residuals

$$ SSR = RSE^2 \cdot df $$

$$ SSR = 3072^2 \cdot 93 $$

**Step 2**

Compute the total sum of squares using the R² reported in the summary
R gives $R^2 = 0.9951$ so

$$ 0.9951 = 1 - \frac{SSR}{SST} $$

Rearrange

$$ \frac{SSR}{SST} = 1 - 0.9951 $$

$$ SST = \frac{SSR}{1 - 0.9951} $$

**Step 3**

Insert $SSR$ from Step 1 and calculate the ratio

$$ R^2 = 1 - \frac{SSR}{SST} $$

This yields

$$ R^2 = 0.9951 $$

To solve this by hand, we use the relationship between the F-statistic and $R^2$. The formula for the F-statistic is defined as

$$F = \frac{\frac{R^2}{p}}{\frac{1 - R^2}{n - p - 1}}$$

where $p$ is the number of predictors and $n - p - 1$ is the degrees of freedom for the residuals. We can rearrange this formula to solve for $R^2$ directly

$$R^2 = \frac{F \cdot p}{F \cdot p + (n - p - 1)}$$

From the R summary output, we have the following values

$$F = 4748$$

$$p = 4$$

$$n - p - 1 = 93$$

Substituting these values into the rearranged equation gives

$$R^2 = \frac{4748 \cdot 4}{4748 \cdot 4 + 93}$$

$$R^2 = \frac{18992}{18992 + 93}$$

$$R^2 = \frac{18992}{19085}$$

$$R^2 \approx 0.995127$$

Rounding to four decimal places, we get

$$R^2 = 0.9951$$

<img src="SymPyBilleder/2025-11-27-11-33-56.png" width="750">

For this question you compare the **full model**

* M1

$$ \text{Church} = \alpha + \beta_1 H3020 + \beta_2 H4058 + \beta_3 H5020 + \beta_4 H8090 + \varepsilon $$

to the **reduced (simpler) model**

$$ \text{Church} = \tilde{\alpha} + \tilde{\beta}_1 H3020 + \tilde{\varepsilon} $$

You test

$$ H_0 : \beta_2 = \beta_3 = \beta_4 = 0 $$

The usual F–test statistic for comparing a reduced and a full model is

$$ F = \frac{(SSE_R - SSE_F) / (df_R - df_F)}{SSE_F / df_F} $$

where

* $SSE_R$ and $df_R$ are the residual sum of squares and residual df for the **reduced** model
* $SSE_F$ and $df_F$ are the residual sum of squares and residual df for the **full** model

In your case

* $SSE_R = 15460700518$
* $SSE_F = 877734461$
* $df_R = 96$
* $df_F = 93$

So the statistic is

$$ F = \frac{(15460700518 - 877734461) / (96 - 93)}{877734461 / 93} = 515 ....$$

This matches the **fourth** option in the list (the fraction with $(96-93)$ in the numerator’s denominator).

```r
# full and reduced models
m_full    <- lm(church ~ H3020 + H4058 + H5020 + H8090, data = exam2023)
m_reduced <- lm(church ~ H3020, data = exam2023)

# sums of squares and dfs
SSE_full <- deviance(m_full)
SSE_red  <- deviance(m_reduced)

df_full <- df.residual(m_full)
df_red  <- df.residual(m_reduced)

# manual F statistic
F_stat <- ((SSE_red - SSE_full) / (df_red - df_full)) / (SSE_full / df_full)

cat("SSE_reduced:", SSE_red, "\n")
cat("SSE_full   :", SSE_full, "\n")
cat("df_reduced :", df_red,  "\n")
cat("df_full    :", df_full, "\n")
cat("F statistic:", F_stat,  "\n")

# check with built-in anova
anova(m_reduced, m_full)
```
