### **Question 2.2**

<img src="SymPyBilleder/2025-11-26-21-07-25.png" width="550">

**Answer:**

We use the standard Bartlett-type chi-square df for testing that the last ($q$) eigenvalues are equal:
$$
\text{df} = \tfrac{1}{2}(q-1)(q+2),
$$

Where $q=k-m$ is the number of smallest eigenvalues being tested. Here $q=7$, so

$$
\text{df}=\tfrac{1}{2}(7-1)(7+2)=\tfrac{1}{2}\cdot 6\cdot 9=27.
$$

**The answer is, therefore, option 4.**
