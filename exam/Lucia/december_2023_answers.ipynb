```python
# We start by writing the code:
educ_vars <- c("H7020","H7025","H7030","H7035","H7039",
               "H7059","H7075","H7080","H7090","H7095")

sport_vars <- c("Cycling","CanoeKayak","Gliding","Hanggliding",
                "HorseRiding","Sailing","Skiing","SurfRafting")

X <- exam2023[, educ_vars]
Y <- exam2023[, sport_vars]

# Canonical correlation:
cc <- cancor(X, Y)
cc$cor
```

Now to test significance we use the Wilks' Lambda test statistic and we write everything in R.

```python
wilks_test <- function(cc, X, Y) {
  n <- nrow(X)
  p <- ncol(X)
  q <- ncol(Y)
  s <- length(cc$cor)

  results <- data.frame(
    Test = paste0("rho_", 1:s, "+"),
    Lambda = NA,
    ChiSq = NA,
    df = NA,
    p_value = NA
  )

  for (j in 1:s) {
    # product over remaining canonical correlations
    lambda_j <- prod(1 - cc$cor[j:s]^2)
    df_j <- (p - j + 1) * (q - j + 1)
    chisq_j <- -(n - 0.5 * (p + q + 3)) * log(lambda_j)
    pval_j <- pchisq(chisq_j, df_j, lower.tail = FALSE)

    results$Lambda[j] <- lambda_j
    results$ChiSq[j] <- chisq_j
    results$df[j] <- df_j
    results$p_value[j] <- pval_j
  }

  return(results)
}

## Run the test:
test_results <- wilks_test(cc, X, Y)
test_results
```

We can count to 4 variables having a $p$-value less than $0.05$.

**The answer is, therefore, option 6.**
