### **Question 6.5**

<img src="SymPyBilleder/2025-11-18-20-24-30.png" width="550">

**Answer:**

We should maybe use $F$-statistic? Because we are testing one variable added to a model, and the correct choice is option 5.

BECAUSE we have two nested discriminant models (reduced = aM2, bM2 while full is = LM2,aM2,bM2). The standard test for “does the extra variable add information?” is an $F$–test formed as a ratio of two mean–square type quantities (incremental explained / residual). So we want the form:
$$
\frac{\text{(sum–of–squares for reduced)}/df_1}{\text{(sum–of–squares for full)}/df_2},
$$
Hence, an $F$–ratio using the two reported scalars ($13.37960$ and $12.76396$) with their appropriate degrees of freedom ($57$ and $56$). That is the correct conversion for the nested–model test in this two–group discriminant setting.

**The answer is, therefore, option 5.**
