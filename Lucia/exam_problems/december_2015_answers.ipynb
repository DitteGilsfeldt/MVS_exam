$$
\rho_{ij|kl} = \frac{\rho_{ij|k} - \rho_{il|k}\rho_{jl|k}}{\sqrt{(1-\rho_{il|k}^2)(1-\rho_{jl|k}^2)}}
$$
$$
= \frac{Cov(X_i, X_j|X_k)}{\sqrt{Var(X_i|X_k)Var(X_j|X_k)}}
$$
Then from the task description we got the matrix:
$$
\Sigma_{Y|X} =  \begin{bmatrix}1 & 1 \\ 1 & 4 \end{bmatrix} - \begin{bmatrix} \frac{13}{192} & \frac{1}{4} \\ \frac{1}{4} & 1 \end{bmatrix} = \begin{bmatrix} \frac{179}{192} & \frac{3}{4} \\ \frac{3}{4} & 3 \end{bmatrix}

$$
V(X_1|X_2 = x_2) = \Sigma_{11} - \Sigma_{12} \Sigma_{22}^{-1} \Sigma_{21}
$$
So instead we have:
$$
V(Y|X) = \Sigma_{YY} - \Sigma_{YX} \Sigma_{XX}^{-1} \Sigma_{XY}
