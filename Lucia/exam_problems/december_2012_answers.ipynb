### **Question 1.4**

<img src="SymPyBilleder/2025-11-19-19-31-16.png" width="550">

**Answer:**

For deleting observation $i$ we use the formula:
$$
\hat\sigma_{(i)}^2 = \frac{(n-p)\hat\sigma^2 - \frac{r_i^2}{1-h_i}\hat\sigma^2}{(n-p)-1}
$$

Where:

> $r_i =$ studentized residual\
> $h_i =$ leverage\
> $\hat\sigma =$ original estimate\
> $\hat\sigma_{(i)} =$ estimate with observation $i$ removed\
> $p =$ number of parameters including the intercept\
> $n =$ number of observations\

The ratio asked for is then:
$$
\frac{\hat\sigma}{\hat\sigma_{(i)}}
$$

We find the relevant SAS output is for MODEL3, and observation $9$ is shown here with student residual (RStudent) $= âˆ’3.100$ and leverage is $= 0.0346$. Furthermore, from MODEL3 ANOVA we have $SSE = 25250$, $df = 58$ so $\hat\sigma^2 = 25250/58$ and $\hat\sigma = 20.86485$.


Also, for MODEL3 we have $n = 60$ where parameters is $p = 2$ (intercept + hightsq so $n - p = 58$ we insert:
$$
\frac{r_i^2}{1-h_i} =  \frac{3.100^2}{1-0.0346}
$$
$$
= \frac{9.61}{0.9654}
$$
$$
\approx 9.956
$$
This is the "correction" term, and now we compute the numerator.
$$
(n-p)\hat\sigma^2 = 58\hat\sigma^2 = \text{SSE} = 25250
$$
$$
=25250 - 9.956\hat\sigma^2
$$
But:
$$
\hat\sigma^2=25250/58=435.34191
$$
So:
$$
9.956\hat\sigma^2 \approx 9.956 \cdot 435.34191 = 4335.3
$$

Thus:
$$
\text{New SSE} \approx 25250 - 4335.3 = 20914.7
$$

Now we can compute $\hat\sigma_{(i)}^2$:
$$
\hat\sigma_{(i)}^2=\frac{20914.7}{57}=366.05
$$
$$
\hat\sigma_{(i)}=\sqrt{366.05}=19.157
$$
FINALLY, the ratio becomes:
$$
\frac{\hat\sigma}{\hat\sigma_{(i)}} = \frac{20.86485}{19.157} \approx 1.089
$$

Closest option is option 4 with $1.085$.

**The answer is, therefore, option 4.**

### **Question 2.3**

<img src="SymPyBilleder/2025-11-19-20-05-41.png" width="550">

**Answer:**

We use this formula:
$$
\rho^2 = \frac{\Sigma_{YX}\Sigma_{XX}^{-1}\Sigma_{XY}}{\sigma_Y^2}
$$
But since we need a variable that is a linear combination we use:
$$
\rho^2(Y,Z) = \frac{(a^T\Sigma_{XY})^2}{V[Y]a^T\Sigma_{XX}a}
$$
Here, $a$ is the vector that makes $Z$ so $a= \begin{bmatrix} \frac{3}{7} \\ \frac{2}{7} \end{bmatrix}$ and we know all values from before or we can find them in the task description. We insert everything:
$$
= \frac{\left(\begin{bmatrix} \frac{3}{7} & \frac{2}{7} \end{bmatrix} \begin{bmatrix} 2 \\ 1 \end{bmatrix}\right)^2}{4 \cdot \begin{bmatrix} \frac{3}{7} & \frac{2}{7} \end{bmatrix} \begin{bmatrix} 4 & 1 \\ 1 & 2 \end{bmatrix} \begin{bmatrix} \frac{3}{7} \\ \frac{2}{7} \end{bmatrix}}
$$
$$
= \frac{\left(\frac{6}{7} + \frac{2}{7}\right)^2}{4 \cdot \begin{bmatrix} \frac{3}{7} & \frac{2}{7} \end{bmatrix} \begin{bmatrix} \frac{14}{7} \\ \frac{7}{7} \end{bmatrix}}
$$
$$
= \frac{( \frac{8}{7} )^2}{4 \cdot \begin{bmatrix} \frac{3}{7} & \frac{2}{7} \end{bmatrix} \begin{bmatrix} 2 \\ 1 \end{bmatrix}}
$$
$$
= \frac{\frac{64}{49}}{4 \cdot \frac{8}{7}} = \frac{\frac{64}{49}}{\frac{32}{7}} = \frac{64}{49} \cdot \frac{7}{32} = \frac{14}{49} = \frac{2}{7}
$$
**The answer is, therefore, option 4.**
