```python
Means <- as.matrix(z$means)
invCov <- solve(pcov)

maha <- matrix(c(rep(0,9)),ncol=3)
mahabis <- matrix(c(rep(0,9)),ncol=3)
for (i in c(1:3)){
  for (j in c(1:3)){
    mu <- Means[i,] - Means[j,]
    maha[i,j] <- t(mu)%*%invCov%*%mu
    mahabis[i,j] <- t(mu)%*%invCov%*%mu - 2*log(prior[j]) 
  }
}
# squared Mahalanobis distances between the 3 Iris types.
# mahabis : Assuming unequal priors
# maha    : Assuming equal priors
```

```python
# Prior probabilities fo the classes
# For setosa | versicolor | virginica :
prior=c(0.2,0.5,0.3)

#linear discriminant analysis
# ~. means all variables are included
z <- lda(species ~ petallength + sepalwidth, data = Iris, prior=prior)

Class_Level_Information = data.frame("Frequency" = z$counts,"Proportion"=z$counts/z$N,"Prior"=z$prior)
print("Class Level Information:")

pcov <- pooled.cov(as.matrix(Iris[, c("petallength", "sepalwidth")]), Iris$species)
Det <- determinant(pcov)
Det$modulus[1]
```

```python
Means <- as.matrix(z$means)
invCov <- solve(pcov)

maha <- matrix(c(rep(0,9)),ncol=3)
mahabis <- matrix(c(rep(0,9)),ncol=3)
for (i in c(1:3)){
  for (j in c(1:3)){
    mu <- Means[i,] - Means[j,]
    maha[i,j] <- t(mu)%*%invCov%*%mu
    mahabis[i,j] <- t(mu)%*%invCov%*%mu - 2*log(prior[j]) 
  }
}
# squared Mahalanobis distances between the 3 Iris types.
# mahabis : Assuming unequal priors
# maha    : Assuming equal priors

zpred <- predict(z)

#Confusion Matrix:
print("Confusion Matrix:")
xtabs(~Iris$species+zpred$class)

#Class Error Rates:
Er_Setosa = 1 - sum(zpred$class == 'setosa' & Iris$species == 'setosa')/50
Er_Versi = 1 - sum(zpred$class == 'versicolor' & Iris$species == 'versicolor')/50
Er_Virgin = 1 - sum(zpred$class == 'virginica' & Iris$species == 'virginica')/50


# CHANGE:
ggplot(Iris, aes(x = petallength, y = sepalwidth, color = species)) +
      geom_point(size = 2) +
      scale_color_manual(values = c("setosa" = "darkgreen",
                                    "versicolor" = "red",
                                    "virginica" = "steelblue")) +
      ggtitle("Petal length vs Sepal width") +
      theme(legend.position = "bottom")

# add the LD1,LD2 to the other columns of Iris dataset.
lda_plot = cbind(Iris,zpred$x)

# CHANGE:
ggplot(lda_plot, aes(x = LD1, y = LD2, color = species)) +
      geom_point(size = 2) +
      scale_color_manual(values = c("setosa" = "darkgreen",
                                    "versicolor" = "red",
                                    "virginica" = "steelblue")) +
      ggtitle("Canonical discriminant functions (LD1 vs LD2)") +
      xlab("LD1") + ylab("LD2") +
      theme(legend.position = "bottom")
```

So we see how the full LDA (from exercise 1) gives better classification with better seperated species, meanwhile, if we only use 2 features, the clusters start overlapping more, as we see with red and blue.

In summary, the consequences are that fewer variables gives less separation power which means higher classification error. Meanwhile, unequal priors leads to decision boundaries shift which leads to systematic bias in classification. Finally, covariance and mean differences still drive the model, but priors control how conservative versus aggressive classification becomes.

##### **3)**

So from the slides, we know that the Bayes rule has something to do with this:
$$
\frac{f_1(\mathbf{x})}{f_2(\mathbf{x})} \ge \frac{L_{21} p_2}{L_{12} p_1}
$$

Then after substitution and simplification, the rule becomes:
$$
\mathbf{x}^T \Sigma^{-1}(\mu_1 - \mu_2) - \frac{1}{2} (\mu_1^T \Sigma^{-1} \mu_1 - \mu_2^T \Sigma^{-1} \mu_2) \ge \ln \left(\frac{L_{21} p_2}{L_{12} p_1}\right)
$$

Basically we have this, and this one has a constant $c$ part, which we must calculate:
$$
\mathbf{x}^T \hat{\Sigma}^{-1} (\hat{\mu}_1 - \hat{\mu}_2) \underbrace{- \frac{1}{2} \hat{\mu}_1^T \hat{\Sigma}^{-1} \hat{\mu}_1 + \frac{1}{2} \hat{\mu}_2^T \hat{\Sigma}^{-1} \hat{\mu}_2}_{\text{The constant threshold part (c)}}
$$
Hence we need:
$$
c = -\frac{1}{2} \hat{\mu}_2^T \hat{\Sigma}^{-1} \hat{\mu}_2 + \frac{1}{2} \hat{\mu}_1^T \hat{\Sigma}^{-1} \hat{\mu}_1
$$
We insert everything in Python.

```python
c = -1/2*mu_2.T*Sigma.inv()*mu_2 + 1/2*mu_1.T*Sigma.inv()*mu_1
c[0]
```

This is the Bayes classification rule assuming equal priors?.

##### **4)**

So for the exercise above we had to use:
$$
c = -\frac{1}{2} \mu_2^T \Sigma^{-1} \mu_2 + \frac{1}{2} \mu_1^T \Sigma^{-1} \mu_1
$$

This constant $c$ is the scalar part of the discriminant function assuming equal prior probabilities ($\pi_1 = \pi_2 = 0.5$) and equal costs.

But now we have to adjust the decision rule since the prior probabilities ($\pi_k$) have changed. The change in priors introduces an extra term into the classification boundary equation. This term is added to the constant $c$ you calculated in exercise 3.

The general decision rule boundary, comparing class 1 and class 2, is where the linear discriminant function (LDF) equals the log-ratio of the priors and costs gives as:

$$
\mathbf{x}^T \mathbf{\Sigma}^{-1} (\boldsymbol{\mu}_1 - \boldsymbol{\mu}_2) = -\underbrace{\frac{1}{2} (\boldsymbol{\mu}_1^T \mathbf{\Sigma}^{-1} \boldsymbol{\mu}_1 +\boldsymbol{\mu}_2^T \mathbf{\Sigma}^{-1} \boldsymbol{\mu}_2)}_{\text{part from ex. 3 (c)}} + \underbrace{\ln \left(\frac{L_{21} \pi_2}{L_{12} \pi_1}\right)}_{\text{new term for ex. 4}}
$$

The new decision rule is to classify $\mathbf{x}$ to group 1 if:
$$
\mathbf{x}^T \mathbf{\Sigma}^{-1} (\boldsymbol{\mu}_1 - \boldsymbol{\mu}_2) > c_{\text{new}}
$$

Where the new threshold $c_{\text{new}}$ is:

$$\mathbf{c}_{\text{new}} = \left(-\frac{1}{2} \mu_1^T \Sigma^{-1} \mu_1 + \frac{1}{2} \mu_2^T \Sigma^{-1} \mu_2\right) + \ln \left(\frac{\pi_2}{\pi_1}\right)$$

Not to find the two $\pi$ values, we know that group 1 (class 10 or whatever) has 493 pixels while the other has 73 pixels. These are our proportions, hence:
$$
\pi_1 = \frac{493}{493 + 73} \approx 0.871 \quad \text{and} \quad \pi_2 = \frac{73}{493 + 73} \approx 0.129
$$
Now we just insert everything in Python.

```python
pi_1 = 0.871
pi_2 = 0.129

c_new = c[0] + ln(pi_1/pi_2)
c_new
```
